## Fetching Real Stock Data from Yahoo Finance
- Tesla
- AAPL
- GOOGL
- RELIANCE.NS

```python
import yfinance as yf
import pandas as pd

# Download data for the above stocks
stock_tsla = yf.download("TSLA", start="2023-01-01", end="2024-01-01") # Tesla
stock_apple = yf.download("AAPL", start="2023-01-01", end="2024-01-01") # Apple
stock_google = yf.download("GOOGL", start="2023-01-01", end="2024-01-01") # Google
stock_reli = yf.download("RELIANCE.NS", start="2023-01-01", end="2024-01-01") # Reliance

# Show the first few rows for each stock
print("Tesla stock data:")
print(stock_tsla.head())
print("\nApple stock data:")
print(stock_apple.head())
print("\nGoogle stock data:")
print(stock_google.head())
print("\nReliance stock data:")
print(stock_reli.head())
```

### Working with one stock at a time
1. Tesla

```python
stock = stock_tsla

# Show basic info
print("\nStock info:")
print(stock.info())

# Show statistical summary
print("\nStatistical summary:")
print(stock.describe())

# Select only 'Close' price column
close_prices = stock["Close"]
print("\nClose prices:")
print(close_prices.head())

# Calculate daily returns
daily_returns = close_prices.pct_change()
print("\nDaily returns:")
print(daily_returns.head())
```

## Visualizing Stock Prices

```python
import matplotlib.pyplot as plt

# Plot stock price trend
stock["Close"].plot(figsize=(10,5), title="Stock Closing Price")
plt.show()
```

```python
import seaborn as sns

# Create a figure
plt.figure(figsize=(10,5))

# Plot closing prices with Seaborn
sns.lineplot(data=stock["Close"], label="Stock Price", color="blue")

plt.title("Stock Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
